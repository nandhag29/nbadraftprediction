```python
def process_name(name):
    if isinstance(name, str):
        for i in range(len(name) - 1, -1, -1):
            if name[i].isupper():
                name = name[:i] + name[i+1:]
            elif name[i].islower():
                break
    return name
```

```python
new_df['Name'] = new_df['Name'].apply(process_name)
```
